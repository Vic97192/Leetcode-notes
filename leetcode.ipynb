# **leetcode 1. Two Sum**

Sol 1 (用兩個軸，畫出方格找組合)
runtime 7758ms memory 18.64MB

```python
nums = [3,2,4]
target = 6
rec = {}
for num in range(len(nums)):
  for value in range(len(nums)):
    if num != value:
      if nums[num] + nums[value] == target:
        rec.update({num:value})

print('[{a},{b}]'.format(a = list(rec.keys())[0], b = list(rec.values())[0]))
```

```python
from re import S
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        rec = {}
        list1 = []
        for num in range(len(nums)):
            for value in range(len(nums)):
              if num != value:
                if nums[num] + nums[value] == target:
                    rec.update({num:value})
                    list1.append(list(rec.keys())[0])
                    list1.append(list(rec.values())[0])
                    return list1

solution = Solution()
a = solution.twoSum(nums, target)
print(a)
```

```python
nums = [2,7,11,15]
target = 9
for pointer in range(len(nums)):
  find = target - nums[pointer] #算2+x=9,然後要找x
  for scan in range(len(nums)):
    if pointer != scan:
      if find == nums[scan]:
        print(f"[{pointer},{scan}]")
```

```python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for pointer in range(len(nums)):
          find = target - nums[pointer]
          for scan in range(len(nums)):
            if pointer != scan:
              if find == nums[scan]:
                return (pointer,scan)

nums = [3,2,4]
target = 6
solution = Solution()
a = solution.twoSum(nums, target)
print(a)
```

# **leetcode 9.Palindrome Number**

Sol 1（用op[::-1]取代for迴圈, 將int轉str）runtime 11ms memory 17.62MB

```python
x = -121
op = list(str(x))
if op == op[::-1]:
  print("yes")
else:
  print("no")
```

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        op = list(str(x))
        if op == op[::-1]:
            return True
        else:
            return False
```

Sol 2(不轉換int to str)

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        half = 0
        if x < 0 or (x != 0 and x % 10 == 0):
            return False
        elif x < 10:
            return True
        while x > half:
            half = 10 * half + x % 10
            x = x // 10

        if half == x or half // 10 == x:
            return True
        else:
            return False
```
